```python
import pandas as pd;
```

```python
series=pd.Series([1,3,4,5,4,34,35]) # indexing is default
series
```

```python
def addten(x) -> int:  # -> return type Optional info, Not Enforced 
    return "abc"
print(addten(3))
```

```python
def multiplywith100(x):
    return x*100

newseries= series.apply(multiplywith100)
print(newseries)
print("old series")
print(series)
```

```python
# lambda and apply together
x=lambda a:a if a>200 else None

print(newseries.apply(x))
y= lambda a :a>200
print(newseries.apply(y))
print(newseries[newseries.apply(y)]) # most important as here if pass the true false array in [] of series it will take as conditional 

```

```python
# best is define a function if lambda is causing trouble
def greaterthan200(x) ->int: # or maybe float
    if x>200:
        return x
    else:
        return None # or np.nan
    
def  greaterthan300(x) ->bool : # removal of row output is boolean
    return x>300

print(newseries.apply(greaterthan200))
print(newseries.apply(greaterthan300))
newseries[newseries.apply(greaterthan300)]
```
