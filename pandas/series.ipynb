```python
import pandas as pd
```

```python
# series in pandas
series1=pd.Series([524,3,3,23,4,23,"sad"])
series1
```

```python
print(series1.index)
print(series1.values)
print(series1.value_counts(sort=False)) # sort on basis of values
```

```python
obj2=pd.Series([53,"fw",6,3],index=['a','f','r','y'])
print(obj2)
obj2.index  
```

```python
for i in obj2.index:
    print(i)

# okay we can also iterate through indexs
```

```python
print(obj2['a'])
print(obj2*3)
condition = obj2.apply(lambda x: isinstance(x, int)) # using lambda function 
obj2=obj2.where(cond=condition)

print(obj2[obj2 > 2])
```
