```python
class Fraction:
    def __init__(self,x,y): # parameterized constructor  requires an input 
        self.num= x
        self.den= y
        
    def __str__(self): # NEW MAGIC METHOD, helps us tell how our object looks like on print function 
        '''
        the super power it has is more how our obj shoud look like when printed
        '''
        return '{}/{}'.format(self.num,self.den)
    
    def __add__(self, other ): # it always need to object na , we will add 2 objects with + symbol then only it will work
        new_num=self.num*other.den + other.num*self.num
        new_den=self.den*other.den
        return '{}/{}'.format(new_num,new_den) # or i could return Fraction(new_x, new_y)  for saving it an object 
    # def __mul__ or __sub__ or __truediv__(self, other) or __floordiv__(self, other
```

```python
fr1= Fraction(52,4)
fr2= Fraction(2,4)
fr1 + fr2  # fr 1 represent self object in function and fr 2 is representing other object in function
```
