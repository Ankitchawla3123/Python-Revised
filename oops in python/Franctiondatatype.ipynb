```python
class Fraction:
    def __init__(self,x,y): # parameterized constructor  requires an input 
        self.num= x
        self.den= y
        
    def __str__(self): # NEW MAGIC METHOD, helps us tell how our object looks like on print function 
        '''
        the super power it has is more how our obj shoud look like when printed
        '''
        return '{}/{}'.format(self.num,self.den)
    
```
