```python
# now we learn change pin
class Atm: 
    def __init__(self): # constructor function  it is a special function 
        # it has one power, we don't have to call the thing
        # constructor automatically calls and execute if we create a class
        self.pin=""
        self.balance=0
        self.menu()
    
    def menu(self):
        userinput=input(""" 
                        hey how are u, select
                        1 for pin create 
                        2 for pin change
                        3 for check balance
                        4 for withdraw
                        5 or anything else for exit
                        """)
        if userinput=="1":
            self.createpin()
        elif userinput=="2":
            self.changepin()
        elif userinput=="3":
            self.check_balance()
        elif userinput=="4":
            self.withdraw()
        else:
            exit()
             
    def createpin(self):
        pin=input("enter your pin")
        self.pin=pin
        
        user_balance=int(input("enter balance" ))
        self.balance=user_balance
        print("pin created successfully")
        self.menu()
    
    def changepin(self):
        oldpin=input("enter old pin")
        if oldpin== self.pin:
            newpin=input("enter new pin")
            self.pin=newpin
            print("pin changed successfully")
        else:
            print("wrong old pin entered")
        
        self.menu()
        
    def check_balance(self):
        print(self.balance)
        self.menu()
    
    def withdraw(self):
        pin=input("enter the pin")
        if pin == self.pin:
            amt= int(input("enter the amount to withdraw"))
            if (amt > self.balance):
                print("bhai paisa kama")
            else:
                self.balance-=amt
                print("balance withdrawn successfully")
        else:
            print("wrong pin entered")
        self.menu()
```
